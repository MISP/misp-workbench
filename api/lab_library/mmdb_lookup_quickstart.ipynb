# mmdb_lookup quickstart

Shows how to call an enrichment module from a notebook. We use
`mmdb_lookup` (MaxMind GeoIP) to resolve an IP to a country/ASN.

Fork this notebook to a personal copy before running.

## 1. List enabled modules

Sanity-check that `mmdb_lookup` is enabled in this workbench.

```python
[m["name"] for m in mwlab.modules() if m["name"] == "mmdb_lookup"]
```

## 2. Look up a single IP

```python
results = mwlab.enrich(value="8.8.8.8", type="ip-dst", module="mmdb_lookup")
print(results)
```

## 3. Geolocate every IP from a search

Pull a small batch of IP attributes and run them through the same module.

```python
ips = mwlab.search_attributes(type="ip-dst", size=5)
for a in ips:
    value = a.get("value")
    print(f"\n── {value} ───────────────")
    results = mwlab.enrich(value=value, type="ip-dst", module="mmdb_lookup")
    print(results)
```
